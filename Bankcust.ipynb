## 1) Print customer Id, customer name and average account_balance maintained by each customer for all  of his/her accounts in the bank.(8 Rows)

```python
query ='''
select t1.customer_id,t1.customer_name,
avg(t2.balance_amount) as avg_balance
from customer t1
left join bank_account_details2 t2
on t1.customer_id = t2.customer_id
group by t1.customer_id,t1.customer_name
limit 8;'''
result =pd.read_sql_query(query,conn)
print(tabulate(result,headers ="keys",tablefmt = 'grid'))
```

2) Print customer_id , account_number and balance_amount , condition that if balance_amount is nil then assign transaction_amount  for account_type = "Credit Card"(4 Rows)

```python
query2 = '''
select t1.customer_id,t1.account_number,t1.balance_amount,

case
  when (t1.balance_amount is null or t1.balance_amount = 0) and t1.account_type = "CREDITCARD"
  then t2.transaction_amount
  else t1.balance_amount
end as adjust_amount

from bank_account_details2 t1
inner join bank_account_transaction1 t2
on t1.account_number = t2.account_number
group by t1.customer_id,t1.account_number'''
result1 =pd.read_sql_query(query2,conn)
print(tabulate(result1,headers ="keys",tablefmt ="grid"))
```

Print account_number and balance_amount , transaction_amount,Transaction_Date from Bank_Account_Details and bank_account_transaction for all the transactions occurred during march,2020 and april, 2020(12 Rows)

```python
query3 = '''
select
 t1.account_number,t1.balance_amount,
 t2.transaction_amount,t2.transaction_date
from bank_account_details2 t1
inner join bank_account_transaction1 t2
on t1.account_number = t2.account_number
where t2.transaction_date between "2020-03-01" and "2020-04-30"
limit 12 ;'''
result3 = pd.read_sql_query(query3,conn)
print(tabulate(result3,headers ="keys",tablefmt ="grid"))
```

```python
import pandas as pd
```

```python
query4 = '''
select
 t1.customer_id,t2.account_number,t2.balance_amount,
 t3.transaction_amount,t3.transaction_date
 from customer t1
 join bank_account_details2 t2
 on t1.customer_id = t2.customer_id
 join bank_account_transaction1 t3
 on t2.account_number = t3.account_number
 where not (strftime('%y',t3.transaction_date) = '2020' and strftime('%m',t3.transaction_date) ='03')
 limit 22;'''
result4 = pd.read_sql_query(query4,conn)
print(tabulate(result4,headers ="keys",tablefmt = "grid"))
```

7) Print all Customer_id, Account_Number, Account_type, and display deducted balance_amount by subtracting only negative transaction_amounts for Relationship_type ="P" ( P - means Primary , S - means Secondary ) .(27 Rows)

```python
query7 ='''
select
t1.customer_id,t1.account_number,t1.account_type,
t1.relationship_type,
t1.balance_amount,t2.transaction_amount,
balance_amount + transaction_amount as deducted_amount
from bank_account_details2 t1
left join bank_account_transaction1 t2
where transaction_amount < 0 and relationship_type = "P"
limit 27;
'''
result7 = pd.read_sql_query(query7 ,conn)
print(tabulate(result7,headers = "keys",tablefmt ="grid"))
```
